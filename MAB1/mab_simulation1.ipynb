```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import bandit
import mab_solver

import importlib

importlib.reload(bandit)
importlib.reload(mab_solver)
```

```python
# Greedy instance
mab = mab_solver.MABSolver()
```

```python
greedy = mab.greedy()
```

```python
# Epsilon greedy instance (eps: 0.1).
epsilon_greedy01 = mab.greedy(epsilon=0.1)
```

```python
# Epsilon greedy instance (eps: 0.01).
epsilon_greedy001 = mab.greedy(epsilon=0.01)
```

```python
# Plot the results
plt.plot(greedy, label='Greedy')
plt.plot(epsilon_greedy01, label='Epsilon-Greedy (=0.1)')
plt.plot(epsilon_greedy001, label='Epsilon-Greedy (=0.01)')
plt.xlabel('Steps')
plt.ylabel('Average Reward')
plt.legend()
plt.show()
```

# Optimistic Initialization

Comparing regular epsilon greedy approach and greedy approach with optimistic initialization. Fixed step size of alpha=0.1

```python
greedy_opt = mab.greedy(initial_estimation=[5]*10, alpha=0.1)
```

```python
epsilon_greedy_fixed = mab.greedy(epsilon=0.1, alpha=0.1)
```

```python
# Plot the results
plt.plot(greedy_opt, label='Greedy with optimistic initial values')
plt.plot(epsilon_greedy_fixed, label='Epsilon-Greedy (=0.1)')
plt.xlabel('Steps')
plt.ylabel('Average Reward')
plt.legend()
plt.show()
```
